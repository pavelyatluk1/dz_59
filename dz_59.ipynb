#### Реалізуйте функцію зчитування зображення з файлу та його відображення за допомогою бібліотеки OpenCV.   Реалізуйте фільтр Собеля для зображення та відображіть його результат.   Реалізуйте медіанний фільтр для зображення та відображіть його результат.   Порівняйте результати фільтрів Собеля та медіанного фільтру та проілюструйте їх відмінності на зображенні.

```python
# імпорт бібліотек
import cv2
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# завантаження зображення за допомогою функції загрузки
def open_file(file):
    image = cv2.imread(file, cv2.IMREAD_COLOR)
    img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis('off')
    plt.show
    
file =  "dacha2.jpg" 
open_file(file)
```

```python
# фільтр функції Собеля і відображення результату (горизонтальна фільтрація)
sobel_1 = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=1)
plt.imshow(sobel_1)
plt.axis('off')
plt.show
```

```python
# фільтр функції Собеля і відображення результату (вертикальна фільтрація)
sobel_2 = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=1)
plt.imshow(sobel_2)
plt.axis('off')
plt.show
```

```python
# використання медіанного фільтру та відображення зображення при ksize = 3
dst = cv2.medianBlur(img, 3)

plt.imshow(dst)
plt.axis('off')
plt.show
```

```python
# використання медіанного фільтру та відображення зображення при ksize = 25
dst = cv2.medianBlur(img, 25)

plt.imshow(dst)
plt.axis('off')
plt.show
```

```python
# порівняння фільтрів Собеля та медіанного фільтра
# cv2.imshow(img, np.hstack((sobel_1, dst)))

im1 = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=1)
im2 = cv2.medianBlur(img, 25)

cv2.imshow(img, np.hstack((im1, im2)))
cv2.waitKey(0);
cv2.destroyAllWindows();
cv2.waitKey(1)
```
